# Discrete and Fast Fourier Transform

The Fourier Coefficients are given by

$$
\tilde{f(}k_n) = \frac{1}{L} \int_{-\frac{L}{2}}^{\frac{L}{2}} e^{-ik_nx}f(x) \operatorname{d}x
$$

First, we initialize Python [This is of little interest]:

```python
import sympy
import numpy as np
import matplotlib.pyplot as plt
from sympy.plotting import *
#sympy.init_printing (use_latex = True)
```

Set wave variables, plot the wave and compute the fourier coefficients stored in the array coeff:

```python
L = 1 #Intervallwidth (=Lenght of the sample in seconds)
sf = 100 #Sampling rate (=Samples per second)
sp = [t/sf for t in range(sf*L)] #set sampling points
f = [2*sympy.floor((t*2)*2 % 2)-1 for t in sp] #set waveform and samples
plt.plot(sp,f)
plt.xlim(0, 1)


coeff = [1/(L*sf)*np.sum([np.exp(-1j*2*np.pi*k*sp[n])*f[n] for n in range(sf*L)]) for k in range(100)]
```

### Now compute the corresponding fourier series up to a certain order and plot them

# CHANGE: magnitude = amplitude, argument = phase offset

```python
x = sympy.Symbol('x')
f = sympy.Function('f')

f1 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(3)])
f2 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(7)])
f3 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(11)])
f4 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(15)])
f5 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(19)])
f6 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(23)])
f7 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(27)])
f8 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(31)])
f9 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(35)])
f10 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(39)])
f11 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(43)])
f12 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(47)])
f13 = np.sum([sympy.exp((2/L)*np.pi*n*x*1j)*coeff[n] for n in range(51)])
p1 = plot(sympy.re(f1), sympy.re(f2), sympy.re(f3),(x,0,1))
plot(sympy.re(f4),(x,0,1))
plot(sympy.re(f5),(x,0,1))
plot(sympy.re(f6),(x,0,1))
plot(sympy.re(f7),(x,0,1))
plot(sympy.re(f8),(x,0,1))
plot(sympy.re(f9),(x,0,1))
plot(sympy.re(f10),(x,0,1))
plot(sympy.re(f11),(x,0,1))
plot(sympy.re(f12),(x,0,1))
plot(sympy.re(f13),(x,0,1))
print(coeff)
```
