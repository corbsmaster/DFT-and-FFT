# Compare DFT and FFT

## Defining DFT function

```python
import numpy as np
def dft(f):
    N=len(f)
    ft=np.zeros(N,dtype=np.complex)
    jft=np.arange(-N/2,N/2,dtype=np.integer)
    nft=np.arange(-N/2,N/2,dtype=np.integer)
    v=int(N/2)
    for n in nft:
        for j in jft:
            ft[n+v]=ft[n+v]+np.exp(-2*np.pi*1j*n*j/N)*f[j+v]
    ft=ft/N
```

## messuring computing time

```python
import time

N=400
n=1

t1_fft=np.zeros(N)
t2_fft=np.zeros(N)
t1_dft=np.zeros(N)
t2_dft=np.zeros(N)

for x in range(1,N):
    f=np.random.random(x)
    t1_fft[x]=time.time()
    i=0
    while i<n:
        np.fft.fft(f)
        i=i+1
    t2_fft[x]=time.time()
    t1_dft[x]=time.time()
    i=0
    while i<n:
        dft(f)
        i=i+1
    t2_dft[x]=time.time()

t_fft=t2_fft-t1_fft
t_dft=t2_dft-t1_dft

print ("ready to plot")
```
